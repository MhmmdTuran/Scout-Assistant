```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

# Chrome tarayıcıyı başlat
driver = webdriver.Chrome(ChromeDriverManager().install())

# Dosyadan oyuncu linklerini oku
dosya_adı = "liste_dosyasi.txt"
with open(dosya_adı, "r") as dosya:
    player_links = dosya.readlines()

# Verileri saklamak için bir liste oluştur
player_data = []

# Her oyuncu linki için dön
for i, player_link in enumerate(player_links):
    # Oyuncunun sayfasına git
    driver.get(player_link.strip())

    # Oyuncu sayfasından veri çek
    # Örnek: Oyuncu adı ve istatistiklerini çekmek için gerekli XPath veya CSS seçicilerini kullanın
    player_name = driver.find_element_by_xpath('XPath_Buraya_Yaz').text
    player_stats = driver.find_element_by_xpath('XPath_Buraya_Yaz').text

    # Verileri bir sözlük olarak saklayabilir veya başka bir şekilde işleyebilirsiniz
    player_data.append({'Oyuncu Adı': player_name, 'Istatistikler': player_stats})

    # Oyuncu sayfasından çık ve bir sonraki oyuncu sayfasına git
    if i < len(player_links) - 1:
        driver.get(player_links[i + 1].strip())

# Tarayıcıyı kapat
driver.quit()

# Elde edilen verileri işlemek için player_data listesini kullanabilirsiniz
# Örnek: Verileri bir Excel dosyasına veya CSV dosyasına yazmak
# Pandas kullanarak verileri CSV'ye yazmak için:
import pandas as pd

df = pd.DataFrame(player_data)
df.to_csv('oyuncu_verileri.csv', index=False)
```
